average function

```python
def average(data): 
    range = len(data)
    sum = sum(data)
    
    return (sum/range)

```

devation function

```python
def deviation(array):
    deviance = []
    mean = average(array)
    for i in array:
        deviance.append(i-mean)
    return deviance
    
```
